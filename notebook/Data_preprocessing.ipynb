## Keeping only english reviews, removing HTML artifacts and strange characters, and normalizing accents: 

English reviews: 

```python
# Most important first: what is in langue?
lang_counts = reviews["langue"].astype(str).str.lower().str.strip().value_counts(dropna=False)
print(lang_counts.head(30))
print("\nUnique langue values:", lang_counts.shape[0])
```

Since the english reviews represent roughly 45% of the dataset, we will only work with english reviews from here on out

```python
import pandas as pd
import re

print("Total reviews:", len(reviews))

reviews_en = reviews[reviews["langue"].astype(str).str.lower().str.strip().eq("en")].copy()

print("English reviews:", len(reviews_en))
print("Percentage kept:", len(reviews_en) / len(reviews))

print("\nLangue distribution (english df):")
print(reviews_en["langue"].value_counts(dropna=False).head(10))
```

HTML artifacts: 

```python
s = reviews_en["review"].fillna("").astype(str)

has_html_tags = s.str.contains(r"<[^>]+>", regex=True)
has_html_entities = s.str.contains(r"&(?:nbsp|amp|quot|lt|gt|#\d+|#x[0-9a-fA-F]+);", regex=True)

print("English rows:", len(reviews_en))
print("HTML tags count:", int(has_html_tags.sum()), "pct:", float(has_html_tags.mean()))
print("HTML entities count:", int(has_html_entities.sum()), "pct:", float(has_html_entities.mean()))

print("\nLangue values inside HTML-tag examples (should all be 'en'):")
print(reviews_en.loc[has_html_tags, "langue"].value_counts(dropna=False))

print("\nExamples with HTML tags:")
display(reviews_en.loc[has_html_tags, ["idplace", "langue", "review"]].head(10))

print("\nExamples with HTML entities:")
display(reviews_en.loc[has_html_entities, ["idplace", "langue", "review"]].head(10))
```

These rates are negligible, so we do not apply aggressive HTML cleaning.
- We only apply minimal normalization (unescape entities + normalize whitespace),
- which is cheap and keeps semantics intact for BM25 + dense embeddings.

```python
import html
import re

# Minimal normalization only (NOT aggressive cleaning)
reviews_en["text"] = reviews_en["review"].fillna("").astype(str)

# Decode entities like &quot; -> "
reviews_en["text"] = reviews_en["text"].apply(html.unescape)

# Remove tags only if present (affects ~10 rows)
reviews_en.loc[has_html_tags, "text"] = reviews_en.loc[has_html_tags, "text"].str.replace(r"<[^>]+>", " ", regex=True)

# Normalize whitespace
reviews_en["text"] = reviews_en["text"].str.replace(r"\s+", " ", regex=True).str.strip()

print("Using English review text with minimal normalization (entities decoded; rare tags removed).")
```

```python
# Verify HTML artifacts are gone (or near-zero) in the cleaned text
s2 = reviews_en["text"].fillna("").astype(str)

tags_after = s2.str.contains(r"<[^>]+>", regex=True)
entities_after = s2.str.contains(r"&(?:nbsp|amp|quot|lt|gt|#\d+|#x[0-9a-fA-F]+);", regex=True)

print("After minimal normalization:")
print("HTML tags count:", int(tags_after.sum()), "pct:", float(tags_after.mean()))
print("HTML entities count:", int(entities_after.sum()), "pct:", float(entities_after.mean()))
```

Strange characters: 

```python
# Replacement char and NBSP
has_replacement = s.str.contains("�", regex=False)
has_nbsp = s.str.contains("\u00A0", regex=False)  # non-breaking space

# Zero-width chars: ZWSP, ZWNJ, ZWJ, BOM
zero_width_re = r"[\u200B\u200C\u200D\uFEFF]"
has_zero_width = s.str.contains(zero_width_re, regex=True)

# Control chars (excluding normal whitespace)
control_re = r"[\x00-\x08\x0B\x0C\x0E-\x1F\x7F]"
has_control = s.str.contains(control_re, regex=True)

print("Contains '�' %:", has_replacement.mean())
print("Contains NBSP (\\u00A0) %:", has_nbsp.mean())
print("Contains zero-width %:", has_zero_width.mean())
print("Contains control chars %:", has_control.mean())

print("\nExamples with strange chars:")
mask = has_replacement | has_nbsp | has_zero_width | has_control
display(reviews_en.loc[mask, ["idplace", "review"]].head(10))
```

Strange/unprintable characters were rare (<0.1%). We applied a minimal cleanup (NBSP, zero-width, control characters, replacement char) to avoid tokenization issues, without altering semantics.

```python
import re

# Work on the cleaned column you already created
# (raw column stays in reviews_en["review"])
reviews_en["text"] = reviews_en["text"].fillna("").astype(str)

# Remove zero-width characters (ZWSP, ZWNJ, ZWJ, BOM)
reviews_en["text"] = reviews_en["text"].str.replace(r"[\u200B\u200C\u200D\uFEFF]", "", regex=True)

# Remove ASCII control characters (excluding normal spaces)
reviews_en["text"] = reviews_en["text"].str.replace(r"[\x00-\x08\x0B\x0C\x0E-\x1F\x7F]", " ", regex=True)

# Replace non-breaking spaces just in case any remain
reviews_en["text"] = reviews_en["text"].str.replace("\u00A0", " ", regex=False)

# Replace the Unicode replacement character � (very rare) with a space
reviews_en["text"] = reviews_en["text"].str.replace("�", " ", regex=False)

# Normalize whitespace again
reviews_en["text"] = reviews_en["text"].str.replace(r"\s+", " ", regex=True).str.strip()

print("Applied minimal strange-character cleanup on reviews_en['text'].")
```

```python
s_check = reviews_en["text"].fillna("").astype(str)

print("After cleanup:")
print("Contains '�' %:", s_check.str.contains("�", regex=False).mean())
print("Contains NBSP (\\u00A0) %:", s_check.str.contains("\u00A0", regex=False).mean())
print("Contains zero-width %:", s_check.str.contains(r"[\u200B\u200C\u200D\uFEFF]", regex=True).mean())
print("Contains control chars %:", s_check.str.contains(r"[\x00-\x08\x0B\x0C\x0E-\x1F\x7F]", regex=True).mean())
```

```python
df_eval=places
english_place_ids = reviews_en["idplace"].dropna().unique()
df_eval = places[places["id"].isin(english_place_ids)].copy()
print("Original places:", len(places))
print("df_eval (EN places only):", len(df_eval))
```

At this stage, we deliberately separate the pipeline into two datasets with different roles. df_models is a lightweight, review-level dataframe containing only (review_id, place_id, review_text, score); it is the working input used by all review-selection scripts because it keeps the raw textual signal in the simplest possible format and makes it easy to implement strategies like Top-K longest/medium reviews, sentiment-balanced sampling (using score if needed), sentence-level filtering, TF-IDF word budgets, and summarization—without being tied to any specific retrieval model. In parallel, df_eval is kept as the full places table (all metadata columns) and is intentionally not polluted with review text: its purpose is to provide a stable ground-truth reference for later evaluation, where relevance can be defined using one or multiple criteria (e.g., same broad type typeR for level-1 relevance, and finer fields such as subcategory/cuisine/hotel type for level-2, or even additional criteria like price range, rating bucket, or location). The next steps are therefore to: 
- (1) implement the different “meaningful review” selection methods that transform df_models into a place-level corpus (place_id → selected_text)
- (2) run the retrieval models (BM25, TF-IDF, dense, hybrid, reranking) on these corpora 
- (3) compute ranking metrics (Precision@K, Recall@K, MRR, HitRate@K, nDCG, plus the level-1/level-2 ranking error) using df_eval as the relevance source, allowing a clean and fair comparison across the 35 method×model combinations.
